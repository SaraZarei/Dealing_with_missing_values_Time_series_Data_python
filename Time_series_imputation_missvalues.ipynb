```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('E:\work\DataSets\\air-quality.csv')
```

```python
df.head(10)
```

```python
df_new=df.set_index('Date')
```

```python
df_new.head()
```

```python
df_new.isnull().sum()/df_new.shape[0]
```

24% of Ozone and 0.04% of Solar were missing

```python
df_new['Ozone'].plot(color='red',marker='o',linestyle='dotted',figsize=(30,5))
```

as we seen there are gaps(miss values) between data points in Ozone feature

<h3>Imputation missing values with Fillna methhod</h3>

<h4>Forward Fill (ffill) strategy</h4>

```python
ffill_imputed=df_new['Ozone'].fillna(method='ffill')
```

```python
ffill_imputed
```

```python

ffill_imputed.plot(color='red',marker='o',linestyle='dotted',figsize=(30,5))
df_new['Ozone'].plot(title='Ozone',marker='o',figsize=(30,5))
```

<h4>backward Fill (bfill) strategy</h4>

```python
ffill_imputed=df_new['Ozone'].fillna(method='bfill')
ffill_imputed.plot(color='red',marker='o',linestyle='dotted',figsize=(30,5))
df_new['Ozone'].plot(title='Ozone',marker='o',figsize=(30,5))
```

<h3>Imputation missing values with Interpolate</h3>

<h4>interpolate with linear strategy</h4>

```python
linear_interpolate=df_new['Ozone'].interpolate(method='linear')
linear_interpolate.plot(color='red',marker='o',linestyle='dotted',figsize=(30,5))
df_new['Ozone'].plot(title='Ozone',marker='o',figsize=(30,5))
```
